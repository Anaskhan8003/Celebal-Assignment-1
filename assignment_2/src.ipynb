```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def print_list(self):
        if self.head is None:
            print("List is empty.")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    def delete_nth_node(self, n):
        if self.head is None:
            print("Cannot delete from an empty list.")
            return

        if n <= 0:
            print("Invalid position. Must be 1 or higher.")
            return

        if n == 1:
            print(f"Deleted: {self.head.data}")
            self.head = self.head.next
            return

        temp = self.head
        for i in range(n - 2):
            if temp is None or temp.next is None:
                print("Index out of range.")
                return
            temp = temp.next

        if temp.next is None:
            print("Index out of range.")
            return

        print(f"Deleted: {temp.next.data}")
        temp.next = temp.next.next

ll = LinkedList()
ll.append(5)
ll.append(10)
ll.append(15)

print("Initial list:")
ll.print_list()

print("\nDeleting 2nd node:")
ll.delete_nth_node(2)
ll.print_list()

print("\nTrying to delete 5th node (out of range):")
ll.delete_nth_node(5)

print("\nDeleting remaining nodes:")
ll.delete_nth_node(1)
ll.delete_nth_node(1)
ll.delete_nth_node(1)  # Should handle empty case

print("\nFinal list:")
ll.print_list()
```
